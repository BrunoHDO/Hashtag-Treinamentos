#### 12. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.

```python
valor_hora = float(input('Quanto você recebe por hora trabalhada: '))
horas_trabalhada = float(input('Quantas horas você trabalhou esse mês: '))
```

#####  Calcule o salário bruto (horas * salario por hora)

```python
salario_bruto = valor_hora * horas_trabalhada
print(f'${salario_bruto}')
```

##### Calcule o desconto do IR (11% do salário bruto)

```python
IR = salario_bruto * 0.11
```

##### Calcule o desconto do INSS (8% do salário bruto)

```python
INSS = salario_bruto * 0.08
```

##### Calcule o desconto do sindicato (5% do salário bruto)

```python
sindicato = salario_bruto * 0.05
```

##### Calcule o salário líquido (salário bruto - descontos)

```python
salario_liquido = salario_bruto - IR - INSS - sindicato
print(f'${salario_liquido}')
```

Considere que a cobertura da tinta é de 1 litro para cada 3 metros quadrados e que a tinta é vendida em latas de 18 litros, que custam R\$ 80,00. Informe ao usuário a quantidades de latas de tinta a serem compradas e o preço total. (para simplificação nesse momento, não se preocupe em arredondar a quantidade de latas a serem compradas - vamos trabalhar isso em breve)

#### 14. Faça um programa que peça o tamanho de um arquivo para download (em MB) e a velocidade de um link de Internet (em Mbps), calcule e informe o tempo aproximado de download do arquivo usando este link (em minutos).

Detalhe: MB significa megabyte, Mb (com b minúsculo) significa megabit. Um megabit é 1/8 de um megabyte. 

```python
#Valor do arquivo
valor_arquivo = float(input('Informe o tamanho do arquivo em MB:'))
#Velocidade da internet
velocidade_internet = float(input('Informe a velocidade da internet em Mbps'))
#Calcule o tempo de download
tempo_download = (valor_arquivo * 8) / velocidade_internet

print(f'Vai demorar {tempo_download} minutos')
```
